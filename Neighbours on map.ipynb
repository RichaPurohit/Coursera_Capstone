```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[43.67145915, -79.5524920661167], zoom_start=10)
# add markers to map
for lat, lng, borough, neigh in zip(Not_NULL_Toronto['latitude'], Not_NULL_Toronto['longitude'], Not_NULL_Toronto['Borough'], Not_NULL_Toronto['Neighbourhood']):
    label = '{}, {}'.format(neigh, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
Downtown_Toronto_data = Not_NULL_Toronto[Not_NULL_Toronto['Borough'] == 'Downtown Toronto'].reset_index(drop=True)
Downtown_Toronto_data.head()
```

```python
address = 'Downtown Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Downtown Toronto are {}, {}.'.format(latitude, longitude))
```
